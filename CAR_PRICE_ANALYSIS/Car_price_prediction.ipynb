# Dataset Summary
The Car Price Dataset contains 10,000 records with 10 attributes detailing used cars and their resale prices. It includes brand, model, year (2000–2023), engine size (1.0L–5.0L), fuel type, transmission, mileage, doors, owner count, and price ($2,000–$18,301).

Newer cars, luxury brands, and lower mileage vehicles generally have higher prices. Automatic, diesel, and hybrid cars also tend to be more valuable. The dataset is useful for price prediction models and market analysis, revealing trends such as depreciation patterns and the rising popularity of hybrid and electric vehicles.

Columns in Dataset
Brand: Car manufacturer (e.g., Kia, Chevrolet, Mercedes).

Model: Specific car model.

Year: Manufacturing year (2000–2023).

Engine_Size: Engine size in liters (1.0L–5.0L).

Fuel_Type: Fuel type (Diesel, Hybrid, Electric, etc.).

Transmission: Transmission type (Manual, Automatic, Semi-Automatic0).

Mileage: Distance traveled (25–299,947 km).

Doors: Number of doors (2–5).

Owner_Count: Number of previous owners (1–5).

 Price: Selling price in USD ($2,000–$18,301)

# Car Price Dataset Insights (Simplified)
### Price Trends by Year
Newer cars (post-2015) have higher prices, while older models are cheaper.

The median price increases steadily for recent cars, reflecting higher market value.

### Mileage and Price Relationship
Higher mileage cars tend to have lower prices due to wear and depreciation.

Lower mileage cars retain value better, especially newer models.

### Engine Size Impact
Cars with larger engines are priced higher, indicating powerful and premium vehicles.

Larger engines often show lower mileage, suggesting they may be preserved or used less frequently.

### Doors and Pricing
4-door cars dominate the market with the widest price range, offering versatility.

2-door cars are generally priced lower due to less practicality.

5-door cars have moderate pricing, often reflecting hatchbacks or compact SUVs.

### Ownership Patterns
Cars with fewer owners generally have higher prices, indicating better maintenance.

Cars with 3+ owners tend to be cheaper due to increased wear and potential issues.

### Fuel Type and Transmission Insights
Hybrid and electric cars are priced higher, reflecting growing demand.

Automatic cars have higher resale values compared to manuals, indicating a preference for convenience.

### Yearly Car Trends
Car production peaked between 2015-2020, resulting in higher availability of newer cars.

Most recent models are 4-door cars, showing market preference for practicality.
