# bar graph for the "class-marks"

```python
import matplotlib.pyplot as plt
import pandas as pd
k=pd.read_csv("class_marks.csv")
k["Q1"]=k["Q1aM4"]+k["Q1bM6"]
k["Q2"]=k["Q2aM6"]+k["Q2bM4"]
k["Q3"]=k["Q3aM5"]+k["Q3bM5"]
k["Q4"]=k["Q4aM3"]+k["Q4bM7"]
k["Q5"]=k["Q5M10"]
k["Q6"]=k["Q6aM4"]+k["Q6bM6"]
a = k.groupby("Q1")["Total"].sum()
plt.figure(figsize=[10, 6])
plt.bar(a.index, a.values, color='skyblue', edgecolor='black')
plt.title("Bar Graph of Total Marks by Q1 Scores")
plt.xlabel("Q1 Scores")
plt.ylabel("Total Marks")
plt.xticks(ticks=a.index)
plt.show()
```
